```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os#

#Đặt tên file Excel
excel_filename = "datasetprj_2012.xlsx"
# Đường dẫn tới thư mục data (lùi lên 2 cấp rồi vào data)
data_dir = os.path.abspath(os.path.join(os.getcwd(), "..", "..", "data"))
file_path = os.path.join(data_dir, excel_filename)

if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found: {file_path}")

df = pd.read_excel(file_path)
print(f" File loaded successfully: {file_path}")
print(f"Initial shape: {df.shape}")
print(f"Columns: {list(df.columns)}\n")

#2. Kiểm tra cột 'Order_Demand'
if 'Order_Demand' not in df.columns:
    raise KeyError(" Column 'Order_Demand' not found in the dataset!")

# Ép kiểu về numeric (tránh lỗi khi có ký tự lạ)
df['Order_Demand'] = pd.to_numeric(df['Order_Demand'], errors='coerce')

# Báo số giá trị bị mất do không thể chuyển sang số
na_before = df['Order_Demand'].isna().sum()
print(f" Number of NaN values in Order_Demand before cleaning: {na_before}")

#3. Loại bỏ các giá trị NaN hoặc âm
rows_before = len(df)
df = df.dropna(subset=['Order_Demand'])
df = df[df['Order_Demand'] >= 0]
rows_after_basic = len(df)
print(f" Removed {rows_before - rows_after_basic} rows due to NaN or negative Order_Demand values.")

#4. Đếm tần suất từng giá trị Order_Demand
value_counts = df['Order_Demand'].value_counts().sort_index()
print("\nTop 10 Order_Demand value frequencies:")
print(value_counts.head(10))

#5. Giữ lại các giá trị xuất hiện ít nhất 3 lần
threshold = 3
valid_values = value_counts[value_counts >= threshold].index
df_cleaned = df[df['Order_Demand'].isin(valid_values)]

print(f"\n Total unique Order_Demand values before filtering: {len(value_counts)}")
print(f" Valid values appearing ≥ {threshold} times: {len(valid_values)}")
print(f" Rows after cleaning: {len(df_cleaned)} / {len(df)}")

#6. Lưu file mới
output_path = r"C:\Users\Admin\Documents\Group6-Project-Python4DS-DSEB-66B\Group6-Project-Python4DS-DSEB-66B\data\datasetprj_2012_cleaned.xlsx"
df_cleaned.to_excel(output_path, index=False)
print(f"\n Cleaned file saved to: {output_path}")

#7. Tóm tắt thống kê nhanh
print("\n Descriptive statistics for Order_Demand after cleaning:")
print(df_cleaned['Order_Demand'].describe())

import os
print("Current working directory:", os.getcwd())
```
