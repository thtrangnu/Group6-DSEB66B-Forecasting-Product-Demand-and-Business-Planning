```python
# Read the Excel file
df = pd.read_excel('../../datasetprj.xlsx')

# Remove rows with negative Order_Demand values
df = df[df['Order_Demand'] >= 0]

# Count the frequency of each Order_Demand value
value_counts = df['Order_Demand'].value_counts()
print("The frequency of each Order_Demand value:")
print(value_counts)

# Identify values that appear less than 3 times
threshold = 3
valid_values = value_counts[value_counts >= threshold].index

# Filter the DataFrame to keep only valid values
df_cleaned = df[df['Order_Demand'].isin(valid_values)]

print(f"The number of rows after cleaning: {len(df_cleaned)}")

# Save the cleaned DataFrame to a new Excel file
df_cleaned.to_excel('datasetprj_cleaned.xlsx', index=False)
print("Saved file: datasetprj_cleaned.xlsx")
```
